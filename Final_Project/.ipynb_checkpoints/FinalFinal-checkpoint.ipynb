```python
# PART TWO: NETWORKX ANALYSIS
```

```python
# Assigning Nodes and Edges for NetworkX
```

```python
G = nx.Graph()

r = range(data.shape[0])

for i in r:
    G.add_node(i)
    print("node created for location", i)
```

```python
# This cell is a concept test

for i in r:
    for p in r:
        print(i,p,distances[i][p])
```

```python
for i in r:
    for p in r:
        if distances[i][p] == 0:
            print("self-referential, no node created")
        else:
            G.add_edge(i,p, length = (1/distances[i][p]))
            print("node created from", i, "to", p, "of length", distances[i][p])
```

```python
G.edges()
```
